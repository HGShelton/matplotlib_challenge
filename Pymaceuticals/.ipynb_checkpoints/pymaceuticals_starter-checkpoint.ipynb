```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
complete_results = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")

# Display the data table for preview
complete_results.head()
```

```python
# Checking the number of mice.
mouse_count = complete_results["Mouse ID"].nunique()
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. - help from Justin, tutor, with bracket placement for duplicate code. 
duplicate_mice = complete_results[complete_results.duplicated(subset=["Mouse ID","Timepoint"], keep=False)]
duplicate_mice 
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID. - help from Justin (tutor) with identifying the type of function 
# needed to run the code
clean_results = complete_results[complete_results["Mouse ID"] != "g989"]
clean_results.head()
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count = clean_results["Mouse ID"].nunique()
mouse_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regime

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
mean_volume = clean_results.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median_volume = clean_results.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
var_volume = clean_results.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
stdev_volume = clean_results.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
sem_volume = clean_results.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

volume_stats = pd.DataFrame({"Mean Tumor Volume": mean_volume,
                            "Median Tumor Volume": median_volume,
                            "Tumor Volume Variance": var_volume,
                            "Tumor Volume Std. Dev.": stdev_volume,
                            "Tumor Volume Std. Err.": sem_volume
                            })
volume_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution) - used Xpert Learning Assistant to define and show 
# code for the aggregate method

# Using the aggregation method, produce the same summary statistics in a single line
agg_volume = clean_results.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(['mean', 'median', 'var', 'std', 'sem'])
agg_volume
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = clean_results.groupby('Mouse ID')['Timepoint'].max()
max_timepoint

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merge_results = pd.merge(clean_results, max_timepoint, on=['Mouse ID', 'Timepoint'])
merge_results
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    treatment_data = merge_results[merge_results['Drug Regimen'] == treatment]

    # add subset - Worked with peer (Zach) to identify code and fix errors
    tumor_volumes = treatment_data['Tumor Volume (mm3)']
    tumor_vol_data.append(tumor_volumes)
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_volumes.quantile([.25,.5,.75])
    Q1 = quartiles[0.25]
    Q3 = quartiles[0.75]
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    potential_outliers = treatment_data[(treatment_data['Tumor Volume (mm3)'] < lower_bound) | 
                                        (treatment_data['Tumor Volume (mm3)'] > upper_bound)]['Tumor Volume (mm3)']
    print(f"{treatment}'s potential outliers: {potential_outliers}")    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
# Worked with peer (Zach) to identify code and fix errors
fig1, ax1 = plt.subplots()
ax1.set_ylabel('Final tumor Volume (mm3)')
ax1.boxplot(tumor_vol_data, flierprops=dict(marker='o', markerfacecolor='red', markersize=10))
ax1.set_xticklabels(treatments)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
# Used Xpert Learning Assistant to write capomulin_data code and how to extract 'Timepoint' and 'Tumor Volume (mm3)' 
# data points to create the line graph 
mouse_id = 'l509'
capomulin_data = clean_results[(clean_results['Mouse ID'] == mouse_id) & (clean_results['Drug Regimen'] == 'Capomulin')]

plt.plot(capomulin_data['Timepoint'], capomulin_data['Tumor Volume (mm3)'], color='b', linestyle='-')
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title(f'Capomulin treatment of mouse {mouse_id}')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
cap_data = clean_results[(clean_results['Drug Regimen'] == 'Capomulin')]
avg_vol = cap_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
mouse_weight = cap_data.groupby('Mouse ID')['Weight (g)'].mean()

plt.scatter(mouse_weight, avg_vol, color='b')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

slope, int, r, p, std_err = st.linregress(mouse_weight, avg_vol)
fit = slope * mouse_weight + intercept

plt.scatter(mouse_weight, avg_vol, color='b')
plt.plot(mouse_weight, fit, color='r', linestyle='-')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()

print(f"The correlation coefficient between weight and the average tumor volume is {round(st.pearsonr(avg_vol,mouse_weight)[0],2)}")
```
